### Creating the LinearRegression Class From Scratch

```python
class LinearRegression:
    def __init__(self):
        pass
    
    def fit(self, X_Array, Y_Array):
        X_1 = np.ones(shape= (y.shape[0],1))
        X_2 = np.array(X_Array)
        X = np.hstack((X_1, X_2))
        XMat = X
        YMat = np.array(Y_Array)
        xTx = np.matmul(XMat.T, XMat)
        if np.linalg.det == 0:
            print('This matrix cannot do inverse')
        self.Ws = np.matmul(np.linalg.inv(xTx), np.matmul(XMat.T, YMat)) 
        
    @property
    def coef_(self):
        return self.Ws[0]
    
    @property 
    def intercept_(self):
        return self.Ws[1:]
    
    def predict(self, X):
        X_ = np.array(X)
        y = np.sum([np.sum(self.intercept_ * X_), self.coef_])
        return y   
```

```python
model = LinearRegression()
```
